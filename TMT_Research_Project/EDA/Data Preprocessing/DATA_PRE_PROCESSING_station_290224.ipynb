```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read CSV file into a DataFrame
data1 = pd.read_csv('StationsRoute54_Hourly.csv')

# Display the DataFrame
#print(df.head())
df=data1.copy()
```

```python
df.columns
```

```python

# Assuming 'df' is your DataFrame
df['Date_Time'] = pd.to_datetime(df['Date_Time'])  # Convert to datetime
df = df.sort_values(by='Date_Time')  # Sort by 'Date_Time'
df
```

```python
# Filter the DataFrame
filtered_df = df[(df['SourceId'] == 17) & (df['DestinationId'] == 11)]
filtered_df
```

```python
# Convert 'Date_Time' column to datetime
filtered_df['Date_Time'] = pd.to_datetime(filtered_df['Date_Time'])

# Filter records with years 2022 and 2023
filtered_df = filtered_df[filtered_df['Date_Time'].dt.year.isin([2022, 2023])]
filtered_df
```

```python
filtered_df.drop(['Unnamed: 0','RouteNumber','Hour','SourceId','DestinationId'], axis=1, inplace=True)
filtered_df
```

```python
filtered_df=filtered_df.reset_index(drop=True)
filtered_df
```

```python
#Fill Missing Timeslot


# Convert 'timeslot' to datetime format if it's not already
filtered_df['Date_Time'] = pd.to_datetime(filtered_df['Date_Time'])

# Generate a range of all possible time slots
start_time = filtered_df['Date_Time'].min()
end_time = filtered_df['Date_Time'].max()
all_possible_time_slots = pd.date_range(start=start_time, end=end_time, freq='1H')

# Create a DataFrame with all possible time slots
all_time_slots_df = pd.DataFrame({'Date_Time': all_possible_time_slots})
all_time_slots_df
```

```python


# Merge the original DataFrame with the DataFrame containing all time slots
merged_df = all_time_slots_df.merge(filtered_df, on='Date_Time', how='left')

# Fill NaN values in 'count' with 0
merged_df['total_count'].fillna(0, inplace=True)

# If needed, you can sort the DataFrame by 'Date_Time'
merged_df.sort_values(by='Date_Time', inplace=True)

# Now, merged_df contains all time slots with count, and is_holiday values based on data1
merged_df
```
