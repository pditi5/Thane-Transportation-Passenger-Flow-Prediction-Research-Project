```python
df = pd.read_csv('/kaggle/input/final-df-stl-verified-all-station-pairs/df_final_stl_verified_all_sp.csv')
print(df)
```

## taking one station pair = Voltas_Gate_Thane_Station_West

```python
keep_station_pair=['Voltas_Gate_Thane_Station_West']


df = df[df['Station_Pair'].isin(keep_station_pair)]
print(df)
```

```python
df = df.drop(columns=['Station_Pair','Sid', 'Did'])
print(df)
```

```python
df_seasonal = df.drop(columns=[ 'Passenger_Count', 'Trend', 'Residual'])
print(df_seasonal)
```

```python
# Create df_trend
df_trend = df.drop(columns=['Passenger_Count', 'Seasonal', 'Residual'])
print(df_trend)
```

```python
# Create df_residual
df_residual = df.drop(columns=['Passenger_Count', 'Seasonal', 'Trend'])
print(df_residual)
```

```python
# Save the dataframes to CSV files
df_seasonal.to_csv('/kaggle/working/df_seasonal.csv', index=False)
df_trend.to_csv('/kaggle/working/df_trend.csv', index=False)
df_residual.to_csv('/kaggle/working/df_residual.csv', index=False)
```
